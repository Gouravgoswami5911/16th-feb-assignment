```python
#Question1: What is a database? Differentiate between SQL and NoSQL databases.




"""A database is a collection of organized data that can be easily accessed, managed, and updated. It is an essential part of 
modern computer applications and systems, where data is often the most valuable asset. Databases can be used for various 
purposes, such as storing customer information, product data, financial records, and more.

There are two main types of databases: SQL (relational) and NoSQL (non-relational) databases. The main difference between these
    two types of databases is in the way they store and retrieve data.

SQL databases use a structured approach to store data, where data is organized into tables with predefined columns and rows. 
SQL databases use the SQL language (Structured Query Language) to manage data, which is a standardized language used by most 
relational databases. SQL databases are known for their strong consistency, robustness, and ability to handle complex 
relationships between data.

On the other hand, NoSQL databases use an unstructured approach to store data, where data is stored in a flexible, schema-less
format, such as key-value pairs, document-based, or graph-based data models. NoSQL databases use a variety of languages to 
manage data, including JavaScript, JSON, and XML. NoSQL databases are known for their flexibility, scalability, and ability to 
handle large volumes of unstructured data.

Overall, the choice between SQL and NoSQL databases depends on the specific needs of the application or system being developed.
SQL databases are better suited for applications that require strong consistency, data integrity, and complex relationships 
between data. NoSQL databases are better suited for applications that require scalability, flexibility, and the ability to 
handle unstructured data."""



```

```python
#Question2: What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.




"""DDL stands for Data Definition Language, which is a subset of SQL used to define and manipulate the structure of database 
objects, such as tables, indexes, and constraints. DDL statements are used to create, modify, or delete database objects.

Here are some examples of DDL statements and their uses:

CREATE - Used to create new database objects, such as tables, views, indexes, and stored procedures.
For example, the following SQL statement creates a new table called "users" with three columns - "id", "name", and "email":
    
    
CREATE TABLE users (
   id INT PRIMARY KEY,
   name VARCHAR(50),
   email VARCHAR(100)
);
    
DROP - Used to remove database objects, such as tables, views, indexes, and stored procedures.
For example, the following SQL statement drops the "users" table:
    
    
DROP TABLE users;




ALTER - Used to modify the structure of existing database objects, such as tables, views, and indexes.
For example, the following SQL statement adds a new column called "phone" to the "users" table:
    
ALTER TABLE users ADD phone VARCHAR(20);



TRUNCATE - Used to delete all the data from a table while keeping its structure intact.
For example, the following SQL statement removes all the data from the "users" table:
    
TRUNCATE TABLE users;"""
    
    
```

```python
#Question3: What is DML? Explain INSERT, UPDATE, and DELETE with an example.



"""DML stands for Data Manipulation Language, which is a subset of SQL used to manipulate the data within a database. DML 
statements are used to insert, 
update, and delete data from tables within a database.

Here are some examples of DML statements and their uses:

INSERT - Used to insert new data into a table.
For example, the following SQL statement inserts a new row into a table called "employees" with values for the "id", "name", 
"age", and "salary" columns:
    
    
INSERT INTO employees (id, name, age, salary)
VALUES (1, 'John Smith', 35, 50000);



UPDATE - Used to modify existing data in a table.
For example, the following SQL statement updates the "salary" column in the "employees" table for the row with an "id" of 1:
  

UPDATE employees SET salary = 60000 WHERE id = 1;



DELETE - Used to delete data from a table.
For example, the following SQL statement deletes the row from the "employees" table with an "id" of 1:
    
    
DELETE FROM employees WHERE id = 1;"""
    
```

```python
#Queston4: What is DQL? Explain SELECT with an example.


"""DQL stands for Data Query Language, which is a subset of SQL used to retrieve data from a database. DQL statements are used 
to 
query data from one or more tables within a database.

The most common DQL statement is SELECT, which is used to retrieve data from one or more tables based on specified criteria.

Here is an example of a SELECT statement:
    
    
SELECT * FROM customers WHERE country = 'USA';



This statement retrieves all the columns and rows from the "customers" table where the "country" column is equal to "USA". 
The "*" 
symbol specifies that all columns should be retrieved. The "FROM" keyword specifies the table to query, and the "WHERE" keyword
specifies the condition to filter the results.

Here is another example of a SELECT statement that retrieves specific columns and sorts the results in ascending order based 
on the "last_name" column:
    
    
SELECT first_name, last_name, email
FROM customers
ORDER BY last_name ASC;



This statement retrieves only the "first_name", "last_name", and "email" columns from the "customers" table, and sorts the 
results in ascending order based on the "last_name" column. The "ORDER BY" 
keyword is used to specify the sorting order, and the "ASC" keyword specifies the ascending order."""
```

```python
#Question5:  Explain Primary Key and Foreign Key.



"""Primary key and foreign key are both important concepts in relational databases that define relationships between tables.

A primary key is a column or set of columns in a table that uniquely identifies each row in that table. This means that each 
row in the table must have a unique value for the primary key column(s). Primary keys are used to enforce data integrity and 
help ensure that the data in the table is accurate and consistent. In most databases, primary keys are also used as the basis
for creating relationships between tables.

For example, consider a database with two tables - "customers" and "orders". The "customers" table might have a primary key 
column called "customer_id", which uniquely identifies each customer. The "orders" table might have a foreign key column called 
"customer_id", which references the "customer_id" column in the "customers" table. This creates a relationship between the two 
tables, where each order is associated with a specific customer.

A foreign key is a column in a table that refers to the primary key of another table. This creates a relationship between the 
two tables and allows for data to be joined between the tables. The values in the foreign key column must match the values in 
the primary key column of the related table.

Using the example above, the "orders" table would have a foreign key column called "customer_id" that references the primary 
key column "customer_id" in the "customers" table. This would ensure that each order is associated with a valid customer in 
the "customers" table.

In summary, primary keys and foreign keys are both important concepts in relational databases. Primary keys are used to
uniquely identify each row in a table and enforce data integrity, while foreign keys are used to create relationships between 
tables by referencing the primary key of another table. Understanding primary keys and foreign keys is important for designing
and managing relational databases."""



```
